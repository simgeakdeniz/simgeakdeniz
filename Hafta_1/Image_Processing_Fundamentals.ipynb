```python
# Bir görüntünün yüksekliği satır sayısıyla,
# genişliği ise sütun sayısıyla doğru orantılıdır!!!
```

```python
# Bir görüntü (normal şartlarda) 0 ile 255 arasında 256 adet tamsayıyla ifade
# edilir.
# 0, siyah rengine; 255 ise beyaz rengine karşılık gelir.
# RGB kanalında siyah rengi: [0,0,0] ile ifade edilir
# RGB kanalında beyaz rengi: [255,255,255] ile ifade edilir
# Gri rengi, siyah ve beyazın eşit oranda karıştırılmasıyla oluşuyorsa
# aynı şekilde 0 ile 255'i toplayıp 2'ye bölersek, elde edeceğimiz sonuç
# bize gri renginin sayısal karşılığını verecektir.
```

```python
# BT.601 standardı: İnsan gözü en iyi yeşil rengini sonra kırmızıyı en sonda
# maviyi ayırt eder.
# BT.601 standardı, üç kanallı RGB görüntüyü tek kanallı gri seviyeye çekmek
# için kullanılan ağırlıklandırma formülüdür.
# Bu formülde en büyük ağırlık yeşil rengine, sonra kırmızıya sonra da maviye
# verilir. Her bir piksel kendi ağırlığıyla çarpılır ve sonunda hepsi toplanarak
# 3 kanaldan tek kanallı bir sonuç elde edilir.
```

```python
gri_img = cv2.cvtColor(my_img,cv2.COLOR_BGR2GRAY)
```

```python
print(f"orijinal görüntünün boyutları ve kanal sayısı = {my_img.shape}")
print(f"Gri görüntünün boyutları ve kanal sayısı = {gri_img.shape}")
```

```python
# İlk olarak BGR kanal sıralamasında okuduğumuz renkli görüntünün
# sırasıyla RGB renk kanalı sıralamasıyla dönüştürülmesi (convert edilmesi)
# ve gri seviye olarak dönüştürülmesi

my_rgb_img = cv2.cvtColor(my_img,cv2.COLOR_BGR2RGB)
my_gray_img = cv2.cvtColor(my_rgb_img, cv2.COLOR_RGB2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7))

# plt.subplot(satır_sayısı, sütun_sayısı, kaçıncı_görüntü_olduğu)
# Örneğin: plt.subplot(3,2,4) yan taraftaki kodun açıklaması aşağıdadır
# satır sayısı x sütün sayısı = gösterilecek toplam görüntü sayısı
# toplam görüntü = 3x2 = 6 görüntü
# sondaki 4 sayısı ise o anki plot'ta hangi görüntünün gösterildiğini ifade eder
# yani 6 görüntü arasından 4. görüntüyü gösterdiğini açıklıyor

plt.subplot(1, 2, 1)
plt.imshow(my_rgb_img)
plt.axis('off')
plt.title("RGB Image")

plt.subplot(1, 2, 2)
plt.imshow(my_gray_img,cmap="gray")
plt.axis('off')
plt.title("Gray Image")
```

```python
print(f"RGB image üzerindeki piksel değeri = {my_rgb_img[500,400]}")
print(f"Gray image üzerindeki piksel değeri = {my_gray_img[500,400]}")
```
